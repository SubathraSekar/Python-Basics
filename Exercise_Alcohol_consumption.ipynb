### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
drinks=pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
```

### Step 4. Which continent drinks more beer on average?

```python
print("Continent drinking more beer on average :",drinks.groupby('continent')['beer_servings'].mean().idxmax())
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby('continent')['wine_servings'].describe()
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
df=drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean()
drinks=drinks.merge(df,on='continent')
drinks.rename(columns={'total_litres_of_pure_alcohol_x':'total_litres_of_pure_alcohol','total_litres_of_pure_alcohol_y':'Alcohol_Mean'},inplace=True)
drinks
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
df=drinks.groupby('continent')['total_litres_of_pure_alcohol'].median()
drinks=drinks.merge(df,on='continent')
drinks=drinks.rename(columns={'total_litres_of_pure_alcohol_x':'total_litres_of_pure_alcohol','total_litres_of_pure_alcohol_y':'Alcohol_Medain'})
drinks
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
spirit=pd.DataFrame(drinks)
spirit=spirit.groupby('continent')['spirit_servings'].describe().loc[:,['mean','min','max']]
spirit
```
